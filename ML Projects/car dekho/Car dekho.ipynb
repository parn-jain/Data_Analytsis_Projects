```python
finaldata = data[data['selling_price']<3500000]
finaldata
```

```python
sns.boxplot(y ='km_driven',data = finaldata)
finaldata = finaldata[finaldata['km_driven']<2000000]
```

```python
le = LabelEncoder()
finaldata['fuel'] = le.fit_transform(finaldata['fuel'])
finaldata['seller_type'] = le.fit_transform(finaldata['seller_type'])
finaldata['transmission'] = le.fit_transform(finaldata['transmission'])
finaldata['owner'] = le.fit_transform(finaldata['owner'])
```

```python
# MLdata = data.copy()
finaldata.drop(columns=['name'],inplace = True)
X = finaldata.drop(columns=['selling_price'])
Y = finaldata[['selling_price']]
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 1)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x_train,y_train)
```

```python
regp = reg.predict(x_test)
regp2 = np.array(regp).reshape(-1)
```

```python
plt.plot(x_test,regp2,color = 'red')
```

```python
plt.plot(x_test,y_test,color = 'red')
```

```python
new = pd.concat([x_test.km_driven.reset_index(drop = True),pd.Series(regp2)],axis = 1)
new.sort_values(by = 'km_driven',inplace = True)
new
```
